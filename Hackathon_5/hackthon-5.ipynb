```python
key_numerical_feature = ['Age', 'Chol', 'MaxHR']
plt.figure(figsize=(15, 5))
for i, feature in enumerate(key_numerical_feature):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='AHD', data=df)
plt.title('Distribution of Target Variable')
plt.xlabel('AHD')
plt.ylabel('Count')
plt.show()
```

```python
df.columns
```

```python
categorical_features = ['Sex', 'ChestPain_nonanginal', 'ChestPain_nontypical', 'ChestPain_typical', 'Fbs', 'ExAng']


plt.figure(figsize=(15, 5))
for i, feature in enumerate(categorical_features):
    plt.subplot(1, len(categorical_features), i + 1)
    sns.countplot(x=feature, hue='AHD', data=df)
    plt.title(f'Relationship between Target  and {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
numerical_features_2 = ['Age', 'Chol', 'MaxHR']
plt.figure(figsize=(15, 5))
for i, feature in enumerate(numerical_features_2):
    plt.subplot(1, len(numerical_features_2), i + 1)
    sns.boxplot(x='AHD', y=feature, data=df)
    plt.title(f'Relationship between Target and {feature}')
    plt.xlabel('AHD (0=No, 1=Yes)')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

```python
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
features_for_correlation = ['Chol', 'Age', 'RestBP']
correlation_subset = df[features_for_correlation].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_subset, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Cholesterol, Age, and Resting Blood Pressure')
plt.tight_layout()
plt.show()
```

```python
gender_mapping = {1: 'M', 0: 'F'}
survival_by_gender = df.groupby('Sex')['AHD'].value_counts(normalize=True) * 100
survival_by_gender.index = survival_by_gender.index.set_levels([survival_by_gender.index.levels[0].map(gender_mapping), survival_by_gender.index.levels[1]])
survival_by_gender
```

```python
survival_rate_gender = df.groupby('Sex')['AHD'].agg(lambda x: (x == 0).sum() / len(x) * 100)
survival_rate_gender.index = survival_rate_gender.index.map(gender_mapping)
survival_rate_gender
```

```python
survival_by_chest_pain = df.groupby(['ChestPain_nonanginal', 'ChestPain_nontypical', 'ChestPain_typical'])['AHD'].value_counts(normalize=True) * 100
survival_rate_chest_pain = df.groupby(['ChestPain_nonanginal', 'ChestPain_nontypical', 'ChestPain_typical'])['AHD'].agg(lambda x: (x == 0).sum() / len(x) * 100)
survival_rate_chest_pain
```

```python
bins = [df['MaxHR'].min() - 1, 100, 120, 140, 160, 180, df['MaxHR'].max() + 1]
labels = ['<100', '101-120', '121-140', '141-160', '161-180', '>180']
df['MaxHR_Bin'] = pd.cut(df['MaxHR'], bins=bins, labels=labels, right=False)
survival_by_maxhr = df.groupby('MaxHR_Bin',observed=False)['AHD'].value_counts(normalize=True) * 100
survival_by_maxhr
```
